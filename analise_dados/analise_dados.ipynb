Passo 1 - Importar as bibliotecas que serão utilizadas no código

```python
import pandas as pd
import plotly.express as px
```

Passo 2 - Importar a base de dados para o Python para que possam ser feitas as primeiras correções.

```python
tabela = pd.read_csv('telecom_users.csv')
```

Passo 3 - Procurar possíveis erros de formatação e eliminar informações não utilizáveis na análise

```python
display(tabela)
tabela = tabela.drop(['Unnamed: 0'], axis=1)
```

```python
tabela = tabela.dropna(axis=1,how='all')
display(tabela)
```

```python
display(tabela.info())
tabela['TotalGasto'] = pd.to_numeric(tabela['TotalGasto'], errors='coerce')
tabela = tabela.dropna(axis=0, how='any')
display(tabela.info())
```

4º Passo - Fazer uma análise inicial para confirmar a porcentagem de clientes que estão cancelando.

```python
display(tabela['Churn'].value_counts(normalize=True).map("{:.1%}".format))
```
